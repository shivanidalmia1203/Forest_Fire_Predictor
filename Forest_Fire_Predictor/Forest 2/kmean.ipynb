```python
train=pd.read_csv('train.csv')
```

```python
# l, n ,h ,confidence
```

```python
train1=train['latitude']
train2=train['longitude']
train3=train['confidence']
```

```python
con=[train1,train2,train3]
```

```python
train_final=pd.concat(con,axis=1)
```

```python
train_final['index']=range(0,14166)
```

```python
split = train_final.groupby('confidence') 
```

```python
L=split.get_group('l')
N=split.get_group('n')
H=split.get_group('h')
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16,8))
cmap=plt.cm.rainbow
norm = matplotlib.colors.BoundaryNorm(np.arange(0,10,1), cmap.N)
plt.scatter(train_final['longitude'], train_final['latitude'],
            cmap=cmap, norm=norm, s=150, edgecolor='none')
plt.xlabel('Latitude', fontsize=18)
plt.ylabel('Longitude', fontsize=18)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans # Importing KMeans
# Creating an instance of KMeans to find 10 clusters
kmeans_L = KMeans(n_clusters=50,max_iter=10)
kmeans_H = KMeans(n_clusters=25,max_iter=10)
kmeans_N = KMeans(n_clusters=125,max_iter=10)

# Using fit_predict to cluster the dataset
X_l = L[['longitude','latitude']].values
X_h = H[['longitude','latitude']].values
X_n= N[['longitude','latitude']].values
predictions_L = kmeans_L.fit(X_l)
predictions_H = kmeans_H.fit(X_h)
predictions_N = kmeans_N.fit(X_n)
```

```python
pred=kmeans_N.fit_predict(X_n)
```

```python
center_L=pd.DataFrame(predictions_L.cluster_centers_)
center_H=pd.DataFrame(predictions_H.cluster_centers_)
center_N=pd.DataFrame(predictions_N.cluster_centers_)
```

```python
center_L.columns=['longitude','latitude']
center_H.columns=['longitude','latitude']
center_N.columns=['longitude','latitude']
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16,8))
cmap=plt.cm.rainbow
norm = matplotlib.colors.BoundaryNorm(np.arange(0,10,1), cmap.N)
plt.scatter(center_N['longitude'], center_N['latitude'],
            cmap=cmap, norm=norm, s=150, edgecolor='green')
plt.scatter(center_H['longitude'], center_H['latitude'],
            cmap=cmap, norm=norm, s=150, edgecolor='red')
plt.scatter(center_L['longitude'], center_L['latitude'],
            cmap=cmap, norm=norm, s=150, edgecolor='blue')
plt.xlabel('Latitude', fontsize=18)
plt.ylabel('Longitude', fontsize=18)
plt.grid()
plt.show()
```
